# Execute the code below

```
import pandas as pd
import numpy as np
link = "https://raw.githubusercontent.com/WildCodeSchool/data-training-resources/main/quests/weather2019.csv"
df_weather = pd.read_csv(link)
```

# You have now a DataFrame with lots of weather data

```
df_weather.head()
```

# PIVOT TABLE

## It's like in Excel, but with Pandas

There are 2 syntaxes, choose your preferred :
- As a method : YOUR_DATAFRAME.pivot_table()
- As a function : pd.crosstab()

```
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', aggfunc = 'mean')
```

## Create a Pivot table with :
- Months on rows
- the maximum of Max Temperature on values
- Opinion on columns

```
df_weather.pivot_table(columns='OPINION', index='MONTH', values='MAX_TEMPERATURE_C', aggfunc='max', fill_value='No such opinion this month')
```

## Pivot table with :
- Days on rows
- Precipitation on values
- Months on columns


```
df_weather.pivot_table(columns='MONTH', index='DAY', values='PRECIP_TOTAL_DAY_MM')
```

## Pivot table with :
- Months on rows
- the sum of Precipitations on values

**Then display it as a column chart.** (optionnal)

```
to_plot_prec_df = df_weather.pivot_table(index='MONTH', values='PRECIP_TOTAL_DAY_MM', aggfunc='sum')
```

```
import matplotlib.pyplot as plt


to_plot_prec_df.plot(style='b-', kind='bar', legend=True)
```

## Pivot table with :
- Months on rows
- the average of Max Temperature on a first column of values
- the average of Sunhours on a second column of values

```
to_plot_df = df_weather.pivot_table(index='MONTH', values=['MAX_TEMPERATURE_C', 'SUNHOUR'], aggfunc='mean')
to_plot_df
```

# Chart
Using your favorite library (MatPlotLib, Seaborn, Plotly or what you want), draw a chart from the previous pivot table with :
- Temperature in a blue column chart
- Sunhours in a red line plot

Tip : Once your pivot table created, you can help yourself with the `.reset_index()`  method.

```
import matplotlib.pyplot as plt

to_plot_df = to_plot_df.rename(columns={'MAX_TEMPERATURE_C': "Temperature", "SUNHOUR": "Sunhours"})
fig, ax = plt.subplots()


to_plot_df.Temperature.plot(ax=ax, style='b-', kind='bar', legend=True)
to_plot_df_res = to_plot_df.reset_index()
to_plot_df_res.Sunhours.plot(ax=ax, style='r-', kind='line', legend=True, secondary_y=True)
```
