- Which actor is mentioned the most often?


```
import pandas as pd

link = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv'
df = pd.read_csv(link)
df
```

Convert unicode and explode

```
import ast

def convert_list(raw_str: str) -> list[str]:
  return ast.literal_eval(raw_str)

df.actors_list = df.actors_list.apply(convert_list)
df = df.explode('actors_list')
df
```

Getting mode

```
df.actors_list.mode()
```
