### Codind 'translate' function

Just getting unique OPINION names for my convenience

```
df2018.value_counts('OPINION')
```

The function itself

```
def translate(opinion: str) -> str:
  match opinion:
    case 'good':
      return 'gut'
    case 'very bad':
      return 'sehr schlecht'
    case 'bad':
      return 'schlecht'
    case 'not good not bad':
      return 'nicht schlecht nicht gut'
    case 'unknown':
      return 'unbekannt'
    case 'very good':
      return 'sehr gut'
    case _:
      return 'unbekannte Meinung'
```

### Concatenating 2018 and 2019 datasets

```
full_df = pd.concat([df2018, df2019])

if full_df.shape[0] == 730:
  print('Good, all data is consistent')
else:
  print(f'Looks like we lost something, as data is only {full_df.shape[0]} row long')
```

### Translating opinion

```
full_df['OPINION'] = full_df['OPINION'].apply(translate)
full_df.loc[:, ['DATE', 'OPINION']]
```

### Appending MONTH column

Bit of preparations

```
full_df.DATE = pd.to_datetime(full_df.DATE)
```

Making MONTH

```
full_df['MONTH'] = full_df.DATE.apply(lambda x: x.month)
full_df.loc[:, ['DATE', 'MONTH']]
```

### Appending day

```
full_df['DAY'] = full_df.DATE.apply(lambda x: x.day)
full_df.loc[:, ['DATE', 'DAY']]
```

### Appending year

```
full_df['YEAR'] = full_df.DATE.apply(lambda x: x.year)
full_df.loc[:, ['DATE', 'YEAR']]
```

## Drawing plots!

### Preparing data a bit

```
from datetime import datetime

full_df['DAY_OF_YEAR'] = full_df.apply(lambda x: int(x.DATE.strftime('%j')), axis=1)
prep_df2018 = full_df.loc[(full_df.YEAR == 2018)]
prep_df2019 = full_df.loc[(full_df.YEAR == 2019)]
```

### Wind speed comparative histogram

```
from matplotlib import pyplot as plt
fig, ax = plt.subplots()

# histogram is all about frequencies of occurences, so x might be as granular as amount of entries
prep_df2018.plot(ax=ax, y='WINDSPEED_MAX_KMH', kind='hist')
prep_df2019.plot(ax=ax, y='WINDSPEED_MAX_KMH', kind='hist')
ax.legend(["Max day windspeed 2018", "Max day windspeed 2019"]);

plt.xlabel("Wind speed km/h")

plt.show()
```

2018 was definitely windier.

### Precipitation boxplots

```
fig = plt.figure()

fig, ax = plt.subplots()

max_temps = [prep_df2018.PRECIP_TOTAL_DAY_MM, prep_df2019.PRECIP_TOTAL_DAY_MM]

ax.boxplot(max_temps)

ax.set_xticks([1, 2], labels=['PRECIP_TOTAL_DAY_MM_2018', 'PRECIP_TOTAL_DAY_MM_2019'])
plt.show()
```

So in 2019 there were more rainy days (look like median moved up), and there were heavier rains (look at outliers at both boxplots)

### Scatterplots for sunhours by date


```
fig, ax = plt.subplots()


prep_df2018.plot.scatter(ax=ax,x='DAY_OF_YEAR', y='SUNHOUR', c='blue', legend=True)
prep_df2019.plot.scatter(ax=ax,x='DAY_OF_YEAR', y='SUNHOUR', c='red', legend=True)

plt.xlabel("Day of year")
plt.ylabel("Sunhours")

ax.legend(["2018", "2019"]);

plt.show()
```

Looks like sunhours in general are quite the same. Good thing sun is kind of immutable.

```
fig = plt.figure()

fig, ax = plt.subplots()

max_temps = [prep_df2018.MAX_TEMPERATURE_C, prep_df2019.MAX_TEMPERATURE_C]

ax.violinplot(max_temps, showmeans=True, showmedians=True)

ax.set_xticks([1, 2], labels=['Max temperature 2018', 'Max temperature 2019'])

plt.show()
```

The mean and median temperature are approximately the same for both years, however, in 2019 there were days with higher temperature, and in 2018 with lower.
